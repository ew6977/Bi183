# 1) Derive the expression of mutual information and prove that I(X;Y) = I(Y;X), and that I(X,Y) >= 0 

By the law of total probability:
\begin{align}
p(x_i) = \sum_{j} p(x_i,y_j)
\end{align}

We also know:
\begin{align}
H(X) &= -\sum_i p(x_{i})log_{2}p(x_{i}) \\
H(H|Y) &= -\sum_{ij} p(x_{i}, y_{j})log_{2}p(x_{i}| y_{j}) \\
\end{align}

We can derive the equation by substituting in what we know:

\begin{align}
I(X;Y)
&= H(X) - H(X|Y) \\
&=-\sum_i p(x_{i})log_{2}p(x_{i}) + \sum_{ij} p(x_{i}, y_{j})log_{2}p(x_{i}| y_{j}) \\
&= \sum_i p(x_{i})log_{2}\left ( \frac{1}{p(x_{i})}\right ) + \sum_{ij} p(x_{i}, y_{j})log_{2}p(x_{i}| y_{j}) \\
&= \sum_i p(x_{i})log_{2}\left ( \frac{1}{p(x_{i})}\right ) + \sum_{ij} p(x_{i}, y_{j})log_{2}\left ( \frac{p (x_i , y_i)}{p(y_i)} \right )\\
&= \sum_{i} \sum_{j}p(x_{i},y_j)log_{2}\left ( \frac{1}{p(x_{i})}\right ) +  p(x_{i}, y_{j})log_{2}\left ( \frac{p(x_i , y_i)}{p(y_i)} \right ) \\
&=\sum_{ij}p(x_{i},y_j)log_{2}\left ( \frac{p(x_{i}, y_{j})}{p(x_{i}),p(y_i)}\right ) \\
\end{align}



### Prove that I(X;Y) = I(Y;X)
\begin{align}
I(X;Y)
&= \left ( \frac{log(p(X|Y))}{p(X)} \right ) \\
&= \left ( \frac{log(p(X,Y))}{p(X)p(Y)} \right )\\
&= \left ( \frac{log(p(Y|X))}{p(Y)} \right ) = I(Y;X) \\
\end{align}


### Prove that I(X,Y) >= 0 

We know: 
\begin{align}
I(X,Y) =\sum_{ij}p(x_{i},y_j)log_{2}\left ( \frac{p(x_{i}, y_{j})}{p(x_{i}),p(y_i)}\right ) \\
\end{align}

We also know relative entropy:
\begin{align}
D(p||q) = \sum p(x) log ( \frac{p(x)}{q(x)}
\end{align}

Therefore: 
\begin{align}
I(X,Y) = \sum_{ij}p(x_{i},y_j)log_{2}\left ( \frac{p(x_{i}, y_{j})}{p(x_{i}),p(y_i)}\right ) = D(p(X,Y)|| p(X)p(Y)) \\
\end{align}

Jensen's inequality tells us that $D(p||q)\geq 0$:
\begin{align}
-D(p||q) = -E_p \left [ log \frac{p(x)}{q(x)} \right ]\\
= E_p \left [ log \frac{p(x)}{q(x)} \right ] \\
\leq log E_p \frac{p(x)}{q(x)}  \\
= 0\\
\end{align}

Therefore:

\begin{align}
I(X,Y) = D(p(X,Y)|| p(X)p(Y)) \geq 0 \\
\end{align}

### Borrowed UPGMA algorithm to construct trees from: https://github.com/lex8erna/UPGMApy/blob/master/UPGMA.py

```python
# A Quick Implementation of UPGMA (Unweighted Pair Group Method with Arithmetic Mean)

# lowest_cell:
#   Locates the smallest cell in the table
def lowest_cell(table):
    # Set default to infinity
    min_cell = float("inf")
    x, y = -1, -1

    # Go through every cell, looking for the lowest
    for i in range(len(table)):
        for j in range(len(table[i])):
            if table[i][j] < min_cell:
                min_cell = table[i][j]
                x, y = i, j

    # Return the x, y co-ordinate of cell
    return x, y


# join_labels:
#   Combines two labels in a list of labels
def join_labels(labels, a, b):
    # Swap if the indices are not ordered
    if b < a:
        a, b = b, a

    # Join the labels in the first index
    labels[a] = "(" + labels[a] + "," + labels[b] + ")"

    # Remove the (now redundant) label in the second index
    del labels[b]


# join_table:
#   Joins the entries of a table on the cell (a, b) by averaging their data entries
def join_table(table, a, b):
    # Swap if the indices are not ordered
    if b < a:
        a, b = b, a

    # For the lower index, reconstruct the entire row (A, i), where i < A
    row = []
    for i in range(0, a):
        row.append((table[a][i] + table[b][i])/2)
    table[a] = row
    
    # Then, reconstruct the entire column (i, A), where i > A
    #   Note: Since the matrix is lower triangular, row b only contains values for indices < b
    for i in range(a+1, b):
        table[i][a] = (table[i][a]+table[b][i])/2
        
    #   We get the rest of the values from row i
    for i in range(b+1, len(table)):
        table[i][a] = (table[i][a]+table[i][b])/2
        # Remove the (now redundant) second index column entry
        del table[i][b]

    # Remove the (now redundant) second index row
    del table[b]


# UPGMA:
#   Runs the UPGMA algorithm on a labelled table
def UPGMA(table, labels):
    # Until all labels have been joined...
    while len(labels) > 1:
        # Locate lowest cell in the table
        x, y = lowest_cell(table)

        # Join the table on the cell co-ordinates
        join_table(table, x, y)

        # Update the labels accordingly
        join_labels(labels, x, y)

    # Return the final label
    return labels[0]
```

# 3) Given the expression profiles in Q3.xlsx based on read counts: 
## 3A) Compute the t-statistic and the p-value between "Group 1" and "Group 2" for each  gene


```python
#Import the data from the two groups
Grp1 = pd.read_csv('Q3_grp1.csv', sep=',')

Grp2 = pd.read_csv('Q3_grp2.csv', sep=',')
Grp1.loc[0][1:]
```

```python
gene_tstat_pval = []
for i in range(51):
    gene = Grp1.loc[i][0]
    stats = scipy.stats.ttest_ind(Grp1.loc[i][1:], Grp2.loc[i][1:], equal_var = False)
    gene_tstat_pval.append([gene, stats[0], stats[1]])
gene_tstat_pval
pd_stat = pd.DataFrame(gene_tstat_pval)
pd_stat.columns = ['Gene', 'tstat', 'pval']
pd_stat.set_index('Gene', inplace=True)
print(len(pd_stat))
pd_stat
```

## 3C) How many genes are differentially expressed with an FWER of 0.05, 0.1, and 0.2 using the Bonferroni correction, the Holm-Bonferroni correction, and the Hochberg correction?

```python
#Making a Dataframe that only has the p-values
col_to_keep = ['pval']
pval_stat = pd_stat[col_to_keep]
pval_stat = pval_stat.sort_values( by = 'pval', ascending = True).reset_index().drop(['Gene'],1)
pval_stat.index = pval_stat.index +1
pval_stat
```

### Bonferroni correction

```python
#Declare a gene to be differentially expressed if p < 0.05/N
FWER_Bonferroni005 = pval_stat[pval_stat['pval'] < 0.05/51]
FWER_Bonferroni01 = pval_stat[pval_stat['pval'] < 0.1/51]
FWER_Bonferroni02 = pval_stat[pval_stat['pval'] < 0.2/51]
print(len(FWER_Bonferroni005))
print(len(FWER_Bonferroni01))
print(len(FWER_Bonferroni02))
```

There are 39 genes that have p-values < 0.05/51 using the Bonferroni correction

There are 40 genes that have p-values < 0.1/51 using the Bonferroni correction

There are 40 genes that have p-values < 0.2/51 using the Bonferroni correction


### Holm-Bonferroni correction

```python
#Seek smallest index, k, such that pk > 0.05/(N+1-k)
holm_bonferroni005 = pval_stat[pval_stat['pval'] > 0.05/(51+1-pval_stat.index)]
holm_bonferroni01 = pval_stat[pval_stat['pval'] > 0.1/(51+1-pval_stat.index)]
holm_bonferroni02 = pval_stat[pval_stat['pval'] > 0.2/(51+1-pval_stat.index)]
print(holm_bonferroni005)
print(holm_bonferroni01)
print(holm_bonferroni02)
```

There are 40 genes differentially expressed using Holm-Bonferroni expression for all FWER values

### Hochberg Correction

```python
#Seek greatest index k such that pk <= 0.05/(N+1-k)
hochberg005 = pval_stat[pval_stat['pval'] <= 0.05/(51 +1 - pval_stat.index)]
hochberg01 = pval_stat[pval_stat['pval'] <= 0.1/(51 +1 - pval_stat.index)]
hochberg02 = pval_stat[pval_stat['pval'] <= 0.2/(51 +1 - pval_stat.index)]
print(hochberg005)
print(hochberg01)
print(hochberg02)
```

There are 40 genes diffentially expressed using Hochberg correction for all FWER values

## 3D) How many genes are differentially expressed with an FDR of 0.05, 0.1, and 0.2 using the Benjamini-Hochberg correction and the Benjamini-Yekutieli correction?

### Benjamini-Hochberg correction

```python
#Seek the greatest index, k such that pk <= 0.2*k/N
#Declare all genes 1,2,... k to be differentially expressed
bh005 = pval_stat[pval_stat['pval'] <= 0.05*(pval_stat.index)/51]
bh01 = pval_stat[pval_stat['pval'] <= 0.1*(pval_stat.index)/51]
bh02 = pval_stat[pval_stat['pval'] <= 0.2*(pval_stat.index)/51]
print(bh005)
print(bh01)
print(bh02)
```

Using the Benjamini-Hochberg correction:

There are 42 genes that have p-values <= 0.05 x k/N 

There are 44 genes that have p-values <= 0.1 x k/N

There are 44 genes that have p-values <= 0.2 x k/N

### Benjamini-Yekutieli correction

```python
#Seek the greatest index k such that pk <= 0.2*k/FN with F= sum_{j=1}^N 1/j
F = 0
for j in range(1,52):
    F += 1/j
    
print(F)

by005 = pval_stat[pval_stat['pval'] <= 0.05*(pval_stat.index)/(51*F)]
by01 = pval_stat[pval_stat['pval'] <= 0.1*(pval_stat.index)/(51*F)]
by02 = pval_stat[pval_stat['pval'] <= 0.2*(pval_stat.index)/(51*F)]
print(by005)
print(by01)
print(by02)
```

Using the Benjamini-Yekutieli correction:

There are 40 genes that have p-values <= 0.05 x k/FN 

There are 40 genes that have p-values <= 0.1 x k/FN

There are 41 genes that have p-values <= 0.2 x k/FN

## 3E) Compare the numbers of differentially expressed genes of C and D and discuss the observed relations.

You find more differentially expressed genes by using methods that gauruntee a FDR rate (Calculations in D) vs. the number of differentially expressed genes that found using methods that gauruntee a FWER rate. 

FWER corrections change the probability that you will have a false positive at all, but FDR corrections alter the proportion of false positives that are found to true positives. This means that with FWER, you are more likely to reject actually true positives, and with FDR, you are more likely to report false negatives as differentially expressed genes. FWER is a bit more stringent than FDR in this way, because you can be more sure that the genes that are reported differentially expressed *are* actually differentially expressed and not false positives
